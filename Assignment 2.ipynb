### Handling Outliers

```python
data.describe()
```

```python
sns.boxplot(data["age"])
```

```python
q1 = data["age"].describe()["25%"]
q3 = data["age"].describe()["75%"]
```

```python
iqr = q3 - q1
```

```python
l_bondary = q1 - (1.5 * iqr)
```

```python
u_bondary = q3 + (1.5 * iqr)
```

```python
data[data["age"] < l_bondary]
```

```python
(data[data["age"] < l_bondary].shape[0]/data["age"].shape[0])*100
```

#### the lower boundary has around 3% in outliers so it is neglectable

### Data visualisation using seaborn and matplotlib

### 1.Number of males and female vs dead or survived

```python
dmale = data[(data["sex"] == "male")  & (data["alive"] == "no")].shape[0]
dfemale = data[(data["sex"] == "female")  & (data["alive"] == "no")].shape[0]

amale = data[(data["sex"] == "male")  & (data["alive"] == "yes")].shape[0]
afemale = data[(data["sex"] == "female")  & (data["alive"] == "yes")].shape[0]


X = ['dead','survived']
female = [afemale,dfemale]
male = [dmale,amale]
  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, female, 0.4, label = 'Female')
plt.bar(X_axis + 0.2, male, 0.4, label = 'Male')
  
plt.xticks(X_axis, X)
plt.ylabel("Number of passengers")
plt.title("Number of males and female vs dead or survived")
plt.legend()
plt.show()
```

#### 1.The number survived peoples are less than number of dead peaple

#### 2.The maximum number of died passengers are male

###  9.Number of males and female vs dead or survived passengers

```python
smale = data[(data["embark_town"] == "Southampton")  & (data["who"] == "man")].shape[0]
cmale = data[(data["embark_town"] == "Cherbourg")  & (data["who"] == "man")].shape[0]
qmale = data[(data["embark_town"] == "Queenstown")  & (data["who"] == "man")].shape[0]

sfemale = data[(data["embark_town"] == "Southampton")  & (data["who"] == "woman")].shape[0]
cfemale = data[(data["embark_town"] == "Cherbourg")  & (data["who"] == "woman")].shape[0]
qfemale = data[(data["embark_town"] == "Queenstown")  & (data["who"] == "woman")].shape[0]

schild = data[(data["embark_town"] == "Southampton")  & (data["who"] == "child")].shape[0]
cchild = data[(data["embark_town"] == "Cherbourg")  & (data["who"] == "child")].shape[0]
qchild = data[(data["embark_town"] == "Queenstown")  & (data["who"] == "child")].shape[0]


X = ['Southampton', 'Cherbourg', 'Queenstown']
male = [smale,cmale,qmale]
female = [sfemale,cfemale,qfemale]
child = [schild,cchild,qchild]

  
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.1, female, 0.2, label = 'Female')
plt.bar(X_axis + 0.2, male, 0.2, label = 'Male')
plt.bar(X_axis + 0.5, child, 0.2, label = 'child')
  
plt.xticks(X_axis, X)
plt.ylabel("Number of passengers")
plt.title("Number of males and female vs dead or survived")
plt.legend()
plt.show()
```

#### a.The largest number of passengers are from Southampton and with high strength of man and low strength of child
